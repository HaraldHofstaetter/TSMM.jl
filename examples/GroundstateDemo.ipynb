```julia
push!(LOAD_PATH, pwd());
using TSSM
```

```julia
V(x) = x^2
m = SchroedingerReal1D(1024, -30, 30, potential=V, cubic_coupling=390)
```

```julia
psi=wave_function(m)
```

```julia
f(x)=exp(-x^2)
set!(psi, f)
```

```julia
include("groundstate.jl")
```

```julia
groundstate!(psi, extrapolation_order=2)
```

```julia
using PyPlot
```

```julia
x=get_nodes(m);
u=get_data(psi,true)
to_real_space!(psi)
plot(x, u[1:end-2].^2)
```
